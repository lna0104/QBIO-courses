```python
def stigande(x = 2, y = 0, z = 9):
    if x > y:
        tmp = y
        y = x
        x = tmp
    if y > z:
        tmp = z
        z = y
        y = tmp
    if x > y:
        tmp = y
        y = x
        x = tmp
    return [x, y, z]
```

```python
def fakultet(x = 5):
    result = 1
    for i in range(1, x + 1):
        result = result * i
    return result
```
